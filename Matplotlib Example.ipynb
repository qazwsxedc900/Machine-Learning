```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2,100)
plt.plot(x, x**2, label='quadratic')
plt.legend()
plt.show()
```

```python
n = 1024
X=np.random.normal(0,1,n)  #avg=0 std=1
Y=np.random.normal(0,1,n)
T = np.arctan2(Y, X)
plt.scatter(X, Y, s=75, c=T, alpha=0.5) #s=size, c=color, alpha=transparency
```

```python
color = X+Y
cmap = plt.get_cmap('Greys')  #use gray map
norm = plt.Normalize(vmin=-3, vmax=3)  #normalize color
plt.scatter(X, Y, s=75, alpha=0.5, c=color, cmap=cmap, norm=norm)
```

```python
def sinus2d(x, y):
    return np.sin(x) + np.sin(y)

xx, yy = np.meshgrid(np.linspace(0, 2 * np.pi, 5), np.linspace(0, 2 * np.pi, 5))
z = sinus2d(xx, yy)  # Create the image on this grid

plt.imshow(z, origin='lower', interpolation='none')
plt.show()
```
